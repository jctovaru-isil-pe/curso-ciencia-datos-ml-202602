```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ver la lista de datasets que tiene la libreria seaborn
print(sns.get_dataset_names())
```

```python
# Cargar el dataset de ejemplo de Seaborn
df = sns.load_dataset('tips')

# Imprimir las primeras filas
print(df.head(10))
```

```python
# ANALISIS DESCRIPTIVO
# -----------------------------------------------------------------------------------------------
# El objetivo es obtener un resumen rapido de las estadisticas centrales y su estructura (df).

# Resumen estadistico
print(df.describe())

# Crear una composición de múltiples gráficos (Subplots)
fig, ax2 = plt.subplots(1, 1, figsize=(12, 4))

ax2.hist(df['tip'], bins=30, color='red', edgecolor='black')
ax2.set_title("Distribución de Propinas")

plt.show()

# Informacion del dataset
print(df.info())

# Conteo de categorias
print(df["sex"].value_counts())
print(df["smoker"].value_counts())
print(df["day"].value_counts())
```

```python
# Analisis Univariado
# La idea es analizar una variable desde diferentes perspectivas

sns.histplot(df['total_bill'], kde=True, color='red')
plt.title('Distribución del Total de la Cuenta')
```

```python
# Analisis Bivariado
# Analizar la relacion entre dos variables, por ejemplo, si a mayor cuenta (total_bill) hay mayor propina (tip) o que tipo de relaciones

# Relación entre Total de la Cuenta y Propina (Numérica vs Numérica)
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Relación: Cuenta vs Propina')
plt.show()
# Coeficiente de correlacion de Pearson
print("Coeficiente de correlacion de Pearson:", df[['total_bill', 'tip']].corr())
```

```python
# Analisis Multivariado
# Aqui estamos hablando de mas de dos variables

# Relación Cuenta vs Propina, diferenciando por Género y si es Fumador
sns.relplot(x='total_bill', y='tip', hue='sex', style='smoker', data=df)
plt.title('Análisis Multivariado: Cuenta, Propina, Sexo y Fumador')
plt.show()
```

```python
# Mapa de calor de correlaciones (solo variables numéricas)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
