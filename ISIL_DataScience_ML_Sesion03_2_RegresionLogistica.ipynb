```python
#Importar librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Cargar (importar) un dataset - Mercado Inmobiliario en California
california = fetch_california_housing(download_if_missing=True)

df = pd.DataFrame(california.data, columns=california.feature_names)
df['Price'] = california.target
```

```python
# Analisis Descriptivo del dataset
print(df.info())
print(df.describe())
print(df.head(10))
```

```python
# Una nueva variable clasificadora que me indique si la casa es de lujo o no (0 - No, 1 - Si)
precio_medio = np.median(df['Price'])
df['EsDeLujo'] = (df['Price'] > precio_medio).astype(int)


# Definir las caracteristicas (X) y la prediccion (y)
X = df[['MedInc', 'HouseAge', 'AveRooms', 'AveOccup']]
y = df['EsDeLujo']

print(df)
```

```python
# Preparar el entrenamiento de los datos (particionarlos) en datos de entrenamiento y datos de prueba
tam_datos_prueba = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=tam_datos_prueba, random_state=42)

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
# Creacion del modelo - Regresion Logistica
model = LogisticRegression()

# Entrenamiento del modelo con los datos de entrenamiento: X_train, y_train
model.fit(X_train, y_train)
```

```python
# Prediccion de Probabilidades (la salida cruda del modelo)
y_prob = model.predict_proba(X_test)[:, 1]

# Prediccion de la clase (la decision final usando el umbral por defecto de 0.5)
y_pred_clase = model.predict(X_test)

print(y_prob)
print(y_pred_clase)
```

```python
# Nuevas predicciones
datos_nuevos = pd.DataFrame([[5.0, 20, 5, 3]], columns=X.columns)

# Clase y Probabilidad de que sea esa Clase
clase_es_lujo = model.predict(datos_nuevos)[0]
prob_es_lujo = model.predict_proba(datos_nuevos)[0, 1]

# Imprimir los resultados
print("La clase es:", clase_es_lujo)
print(f"La probabilidad de que sea de lujo es: {prob_es_lujo*100: .2f}%")
```
