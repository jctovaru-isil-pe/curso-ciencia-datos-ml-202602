# Importacion de Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
```

# Uso de Pandas

```python
# Colocar la URL del Dataset (direccion)
url = "https://raw.githubusercontent.com/shrikant-temburwar/Wine-Quality-Dataset/refs/heads/master/winequality-white.csv"

# Cargar el Dataset en un DataFrame
df = pd.read_csv(url, sep=';')

# Imprimir
print("Dataset cargado: ", df.shape)

# Imprimir los primeros "n" registros
print(df.head(10))

# Imprimir la informacion descriptiva del dataset
print(df.info())

# Obtener el vector (fila) de los vinos de alta calidad (quality >= 7)
high_quality_alcohol = df[df['quality'] >= 7]['alcohol']
print(high_quality_alcohol)

# Agrupamiento: El equivalente a una tabla dinamica en Excel
df.groupby('quality')[['density', 'pH']].median()

# Obtener la lista del azucar residual
hq_residual_sugar = df[df['quality'] >= 7]['residual sugar']
print(hq_residual_sugar)
```

# Uso de NumPy

```python
print(df['residual sugar'])

df['log_residual_sugar'] = np.log1p(df['residual sugar'])
print(df['log_residual_sugar'])
```

# Uso de MatPlotLib

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.hist(df['residual sugar'], bins=30, color='orange', edgecolor='black')
ax1.set_title('Evolucion del Azucar Residual (original)')

ax2.hist(df['log_residual_sugar'], bins=30, color='skyblue', edgecolor='black')
ax2.set_title('Evolucion del Azucar Residual (transformado a escala)')

plt.show()
```
